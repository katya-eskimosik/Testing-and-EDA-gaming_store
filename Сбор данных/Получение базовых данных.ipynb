```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time

def parse_steam_game(app_id):
    url = f"https://store.steampowered.com/app/{app_id}/"
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}

    try:
        response = requests.get(url, headers=headers)
        if 'agecheck' in response.url:
            return None

        soup = BeautifulSoup(response.content, 'html.parser')

        title = soup.find('div', {'class': 'apphub_AppName'})
        title = title.text.strip() if title else 'N/A'

        developer_block = soup.find('div', {'id': 'developers_list'})
        developers = [a.text.strip() for a in developer_block.find_all('a')] if developer_block else []

        language_table = soup.find('table', {'class': 'game_language_options'})
        supported_languages = []
        if language_table:
            for row in language_table.find_all('tr')[1:]:
                lang_cell = row.find('td', {'class': 'ellipsis'})
                if lang_cell:
                    supported_languages.append(lang_cell.text.strip())

        metacritic_block = soup.find('div', {'id': 'game_area_metascore'})
        metacritic_score = 'N/A'
        if metacritic_block:
            score_element = metacritic_block.find('div', {'class': 'score'})
            if score_element:
                metacritic_score = score_element.text.strip()

        category_block = soup.find('div', {'class': 'glance_tags popular_tags'})
        categories = [a.text.strip() for a in category_block.find_all('a')] if category_block else []

        return {
            'appid': app_id,
            'name': title,
            'developers': '; '.join(developers),
            'supported_languages': '; '.join(supported_languages),
            'metacritic_score': metacritic_score,
            'categories': '; '.join(categories)
        }

    except Exception as e:
        print(f"   {app_id}: {str(e)}")
        return None

def start(ind):
    results = []
    game_ids = df['Game ID'][1000*ind:1000*(ind+1)]

    for i, app_id in enumerate(game_ids, 1):
        print(f"  {app_id}... : {i/len(game_ids)*100:.1f}%")
        game_data = parse_steam_game(app_id)
        if game_data:
            results.append(game_data)
        time.sleep(1.5)

    if results:
        pd.DataFrame(results)[['appid', 'name', 'developers', 'supported_languages', 'metacritic_score', 'categories']].to_csv(f'steam_games_data_{ind}.csv', index=False)
        print(f" {len(results)} ")
    else:
        print("  ")
```

```python
for ind in range(27):
  start(ind)
```
